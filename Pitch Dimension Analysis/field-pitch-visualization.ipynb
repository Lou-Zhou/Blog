```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('per_manager_complete.csv')#read csv, visualization notebook
df.head(20)
```

```python
dflongerdiff = df[["Manager","distance_travelled_home","distance_field_narrower"]]
dflongerdiff = dflongerdiff.dropna()
dflongerdiff["Change in Distance"] = dflongerdiff["distance_field_narrower"] - dflongerdiff["distance_travelled_home"]
print(dflongerdiff["Change in Distance"])
print(np.mean(dflongerdiff["Change in Distance"]))
```

```python
fig, ax = plt.subplots()
bar_color = [('blue' if p > 0 else 'red') for p in dflongerdiff["Change in Distance"]]
ax.barh(dflongerdiff["Manager"], dflongerdiff["Change in Distance"], color = bar_color)

ax.set_xlabel('Distance(m)')
ax.set_title('Passing Distance Change, Wider Field')
plt.xticks(rotation=90)

plt.show()
plt.savefig('longer.png')
```

```python
df_nodiff = pd.read_csv('per-game-2/per_game (1).csv')
df_nodiff = df_nodiff[df_nodiff["Team"] != df_nodiff["home_team_name"]] #if away
```

```python
fielddata = pd.read_csv('fielddata/fielddata.csv')
df_nodiff = df_nodiff.merge(fielddata, left_on = "Team", right_on = "Name")
df_nodiff_1 = df_nodiff[df_nodiff["y_field_length_meters_x"] == df_nodiff["y_field_length_meters_y"]]
df_nodiff_2 = df_nodiff_1[df_nodiff_1["x_field_length_meters_x"] == df_nodiff_1["x_field_length_meters_y"]]
df_nodiff_2 = df_nodiff_2[["Manager", "distance_travelled"]]
df_nodiff_2
```

```python
df_away_no_diff = df_nodiff_2.groupby("Manager").mean()
df_away_no_diff = df_away_no_diff.reset_index()
df_away_no_diff = df.merge(df_away_no_diff, right_on = "Manager", left_on = "Manager", how = 'outer')
df_away_no_diff = df_away_no_diff[:-1]
df_away_no_diff = df_away_no_diff.rename(columns = {"distance_travelled":"distance_field_same"})
df_away_no_diff


```

```python
df_away_same = df_away_no_diff[["Manager","distance_travelled_home","distance_field_same"]]
df_away_same["Change"] = df_away_same["distance_field_same"] - df_away_same["distance_travelled_home"]
df_away_same = df_away_same.dropna()
```

```python
np.mean(df_away_same["Change"])
```

```python
fig, ax = plt.subplots()
bar_color = [('blue' if p > 0 else 'red') for p in df_away_same["Change"]]
ax.barh(df_away_same["Manager"], df_away_same["Change"], color = bar_color)

ax.set_xlabel('Distance(m)')
ax.set_title('Passing Distance Change, Same Field')
plt.xticks(rotation=90)

plt.show()
plt.savefig('longer.png')
```
